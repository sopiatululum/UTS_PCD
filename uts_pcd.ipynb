```python
img = cv2.imread('/content/2010631170033_pas foto.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

- "img = cv2.imread" digunakan untuk membaca sebuah gambar dari file ke dalam program Python menggunakan library OpenCV.

- "gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)", fungsi "cv2.cvtColor" dari library OpenCV digunakan untuk mengubah gambar "img" dari format warna BGR menjadi gambar grayscale. Hasilnya akan disimpan dalam variabel "gray".

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
```

Kode diatas digunakan untuk mendeteksi wajah dalam gambar atau video.
- kita memuat classifier cascade yang telah dilatih sebelumnya untuk mendeteksi wajah. 
- kita menggunakan fungsi detectMultiScale() dari cascade classifier untuk mendeteksi wajah pada gambar atau video.


```python
from google.colab.patches import cv2_imshow

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
cv2_imshow(img)
cv2.waitKey()
```

- menyimpan gambar dengan fungsi imread pada opencv.

- for (x, y, w, h) in faces: - Loop untuk setiap bounding box yang dihasilkan dari deteksi wajah, di mana x dan y adalah koordinat titik pojok kiri atas bounding box, w adalah lebar bounding box, dan h adalah tinggi bounding box.

- cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2) - Menggambar bounding box pada gambar input menggunakan fungsi "rectangle()".
